# 1. Clustering - Part 1 (Kmean Clustering)

```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
setup(data=seedsDataSet)
KMeanClusteringModel1 = create_model('kmeans', num_clusters=3)
KMeanClusteringModel2 = create_model('kmeans', num_clusters=4)
KMeanClusteringModel3 = create_model('kmeans', num_clusters=5)
```

# Conclusion

The K-Means algorithm, when applied to different processing techniques such as normalization and PCA, showed effective clustering with distinct groupings. It performed consistently well across various scenarios, but the quality of clustering improved with data preprocessing, particularly when normalization and PCA were applied.

Hierarchical clustering, while providing a clear tree structure for data points, tended to be sensitive to outliers and produced more coherent clusters only after preprocessing steps like normalization and PCA. Without these, it was less effective than K-Means.

K-Means Shift clustering performed well without needing extensive data preprocessing. It naturally adapted to the shape of the data but had higher computational costs compared to K-Means and Hierarchical clustering.

In summary, K-Means provided a good balance of efficiency and performance with preprocessing, while Hierarchical clustering needed normalization for better results. K-Means Shift offered flexibility with raw data but at a computational cost. Overall, the choice depends on the specific dataset and computational resources, but K-Means with preprocessing appeared to be the most consistent in terms of performance.
